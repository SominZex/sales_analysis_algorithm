### Advanced Analysis and prediction

```python
def sales_over_time_with_all_dates(df, date_col='orderDate', sales_col='totalProductPrice'):
    """
    Plots total sales over time (daily) with all dates shown on the x-axis and data labels for each point.
    """
    # Ensure date column is datetime
    df[date_col] = pd.to_datetime(df[date_col])

    # Create a date range covering all dates in the data
    all_dates = pd.date_range(df[date_col].min(), df[date_col].max())

    # Group by date and sum sales
    daily_sales = df.groupby(df[date_col].dt.date)[sales_col].sum().reset_index()
    daily_sales[date_col] = pd.to_datetime(daily_sales[date_col])

    # Reindex to include all dates, fill missing with 0
    daily_sales = daily_sales.set_index(date_col).reindex(all_dates, fill_value=0).reset_index()
    daily_sales.rename(columns={'index': date_col}, inplace=True)

    # Convert date to string for categorical x-axis
    daily_sales['date_str'] = daily_sales[date_col].dt.strftime('%Y-%m-%d')

    # Plot with all dates as categories
    fig = px.line(
        daily_sales,
        x='date_str',
        y=sales_col,
        title="Total Sales - Daily (All Dates)",
        labels={sales_col: 'Total Sales', 'date_str': 'Date'},
        text=sales_col
    )
    fig.update_traces(textposition='top center')
    fig.update_xaxes(type='category', tickangle=-45, tickmode='array', tickvals=daily_sales['date_str'])
    fig.show(renderer="browser")
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

def rfm_analysis(df, customer_col='customerNumber', date_col='orderDate', sales_col='totalProductPrice'):
    """
    Perform RFM (Recency, Frequency, Monetary) analysis.
    Returns a DataFrame with RFM scores.
    """
    df[date_col] = pd.to_datetime(df[date_col])
    snapshot_date = df[date_col].max() + pd.Timedelta(days=1)
    rfm = df.groupby(customer_col).agg({
        date_col: lambda x: (snapshot_date - x.max()).days,
        'invoice': 'nunique',
        sales_col: 'sum'
    }).rename(columns={date_col: 'Recency', 'invoice': 'Frequency', sales_col: 'Monetary'})

    rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=range(5, 0, -1))
    rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=range(1, 6))
    rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=range(1, 6))
    rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
    rfm['RFM_Score'] = rfm[['R_Score', 'F_Score', 'M_Score']].astype(int).sum(axis=1)
    return rfm

def basket_analysis_first_week(df, invoice_col='invoice', product_col='productName', date_col='orderDate', min_support=0.01, min_confidence=0.3):
    """
    Perform market basket analysis using Apriori algorithm on only the first week's data.
    Returns association rules DataFrame.
    """
    df[date_col] = pd.to_datetime(df[date_col])
    first_date = df[date_col].min()
    last_date = first_date + pd.Timedelta(days=6)
    df_week = df[(df[date_col] >= first_date) & (df[date_col] <= last_date)]

    basket = (df_week
              .groupby([invoice_col, product_col])['quantity']
              .sum().unstack().fillna(0))
    basket = basket.applymap(lambda x: 1 if x > 0 else 0)

    from mlxtend.frequent_patterns import apriori, association_rules
    frequent_itemsets = apriori(basket, min_support=min_support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
    return rules

def sales_time_series(df, date_col='orderDate', sales_col='totalProductPrice', freq='D'):
    """
    Returns a DataFrame with sales aggregated by the specified frequency (D/W/M).
    """
    df[date_col] = pd.to_datetime(df[date_col])
    ts = df.groupby(pd.Grouper(key=date_col, freq=freq))[sales_col].sum().reset_index()
    return ts

def price_elasticity(df, product_col='productName', price_col='sellingPrice', quantity_col='quantity'):
    """
    Estimates price elasticity for each product using log-log regression.
    Returns a DataFrame with elasticity values.
    """
    from scipy.stats import linregress
    elasticity = []
    for product, group in df.groupby(product_col):
        if group[price_col].nunique() > 1:
            x = np.log(group[price_col])
            y = np.log(group[quantity_col])
            slope, _, _, pval, _ = linregress(x, y)
            elasticity.append({'product': product, 'elasticity': slope, 'p_value': pval})
    return pd.DataFrame(elasticity)

def sales_anomaly_detection(df, date_col='orderDate', sales_col='totalProductPrice', z_thresh=3):
    """
    Detects days with anomalously high or low sales using Z-score.
    Returns a DataFrame with anomalies flagged.
    """
    df[date_col] = pd.to_datetime(df[date_col])
    daily_sales = df.groupby(df[date_col].dt.date)[sales_col].sum().reset_index()
    daily_sales['z_score'] = (daily_sales[sales_col] - daily_sales[sales_col].mean()) / daily_sales[sales_col].std()
    daily_sales['anomaly'] = daily_sales['z_score'].abs() > z_thresh
    return daily_sales

def customer_churn(df, customer_col='customerNumber', date_col='orderDate', churn_days=90):
    """
    Flags customers who haven't purchased in the last `churn_days`.
    Returns a DataFrame of churned customers.
    """
    df[date_col] = pd.to_datetime(df[date_col])
    last_purchase = df.groupby(customer_col)[date_col].max().reset_index()
    cutoff = df[date_col].max() - pd.Timedelta(days=churn_days)
    churned = last_purchase[last_purchase[date_col] < cutoff]
    return churned

def basket_metrics(df, invoice_col='invoice', quantity_col='quantity', sales_col='totalProductPrice'):
    """
    Returns average basket size (items per invoice) and average basket value.
    """
    basket = df.groupby(invoice_col).agg(
        basket_size=(quantity_col, 'sum'),
        basket_value=(sales_col, 'sum')
    )
    avg_size = basket['basket_size'].mean()
    avg_value = basket['basket_value'].mean()
    return {'average_basket_size': avg_size, 'average_basket_value': avg_value}

def hourly_and_weekday_patterns(df, date_col='orderDate', time_col='time', sales_col='totalProductPrice'):
    """
    Returns sales aggregated by hour and by day of week.
    """
    df[date_col] = pd.to_datetime(df[date_col])
    df['hour'] = pd.to_datetime(df[time_col], format='%H:%M:%S').dt.hour
    df['weekday'] = df[date_col].dt.day_name()
    hourly = df.groupby('hour')[sales_col].sum().reset_index()
    weekday = df.groupby('weekday')[sales_col].sum().reindex([
        'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
    ]).reset_index()
    return hourly, weekday

def contribution_margin(df, group_col='productName', sales_col='totalProductPrice', cost_col='costPrice', quantity_col='quantity'):
    """
    Calculates contribution margin for each product/category.
    """
    df['total_cost'] = df[cost_col] * df[quantity_col]
    summary = df.groupby(group_col).agg(
        total_sales=(sales_col, 'sum'),
        total_cost=('total_cost', 'sum')
    )
    summary['contribution_margin'] = summary['total_sales'] - summary['total_cost']
    summary['margin_percent'] = 100 * summary['contribution_margin'] / summary['total_sales']
    return summary.reset_index()
```

```python
def basket_analysis_sampled(df, invoice_col='invoice', product_col='productName', date_col='orderDate',
                            min_support=0.001, min_confidence=0.1, top_n_products=20, max_invoices=1000):
    """
    Perform market basket analysis on a sample of data (top N products and max_invoices invoices).
    """
    import pandas as pd
    from mlxtend.frequent_patterns import apriori, association_rules

    df[date_col] = pd.to_datetime(df[date_col])
    top_products = df.groupby(product_col)['quantity'].sum().nlargest(top_n_products).index
    df = df[df[product_col].isin(top_products)]

    sampled_invoices = df[invoice_col].drop_duplicates().sample(n=min(max_invoices, df[invoice_col].nunique()), random_state=42)
    df = df[df[invoice_col].isin(sampled_invoices)]

    basket = (df
              .groupby([invoice_col, product_col])['quantity']
              .sum().unstack().fillna(0))
    basket = basket > 0

    frequent_itemsets = apriori(basket, min_support=min_support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
    return rules
```
